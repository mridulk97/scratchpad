```python
df['English'] = df['English'].str.lower()
df['English'] = df['English'].str.replace('-','_')
df['English'] = df['English'].str.replace(' ','_')
df['English'] = df['English'].str.replace('\'','')
```

```python
df_csv['English'] = df_csv['English'].str.lower()
df_csv['English'] = df_csv['English'].str.replace('-','_')
df_csv['English'] = df_csv['English'].str.replace(' ','_')
df_csv['English'] = df_csv['English'].str.replace('\'','')
```

```python
cub_common_to_scitific_name = {}
for species in cub_species_names:
    cub_common_to_scitific_name[cub_species_names[species]] = df_csv[df_csv['English']==species]['Scientific'].item()
    # print(cub_species_names[species])
    # print(df_csv[df_csv['English']==species]['English'])
```
